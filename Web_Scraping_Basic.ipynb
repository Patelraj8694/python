```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://codewithharry.com"
```

<h3>Get the HTML</h3>

```python
r = requests.get(url)
htmlContent = r.content
```

<h3>Parse the HTML</h3>

```python
soup = BeautifulSoup(htmlContent, 'html.parser')
```

```python
print(soup.prettify())
```

<h3>Html Tree Traversal</h3>

```python
#Title of HTML page

title = soup.title
print(title)
```

```python
print(type(title.string))
print(type(soup))
print(type(title))
```

```python
#All Paragraph
paras = soup.find_all('p')
print(paras)
```

```python
#Get All the links in the page
anchors = soup.find_all('a')

all_links = set()
for link in anchors:
    if(link.get('href') != '#'):
        linktext = "https://codewithharry.com" +link.get('href')
        all_links.add(linktext)
        print(linktext)

```
